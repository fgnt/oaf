# Exercise 3: Linear Prediction

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from ipywidgets.widgets import interact, IntSlider, FloatSlider
```

## a) Source and Predictor

```python
N = 1000000

def source(tap, sigma_u=1, N=1000000):
    u = sigma_u * np.random.randn(N)
    
    x = [0, 0]
    for u_ in u:
        x += [u_ - tap[0] * x[-1] - tap[1] * x[-2]]
    return np.asarray(x[2:])
```

```python
def predictor(w, x):
    d = [0]
    for x_ in x:
        d += [x_ * w[0] + w[1] * d[-1]]
    return np.asarray(d[1:])
```

## b) 

```python
def fun_sigma_u(sigma_x, tap):
    return np.sqrt(sigma_x**2*(1+tap[1]*(tap[0]**2/(1+tap[1])-tap[1])-tap[0]**2/(1+tap[1])))
```

## c) Gradient

```python
def grad_descent(x, a, epsilon=1e-6, step_size=1., num_steps=10000):
    # initial weights
    w = [[0,0]]
    J = []
    
    # optimal weights (for convergence condition)
    covariance_matrix_x = np.cov((x[::2], x[1::2]))
    cross_correlation_vector_dx = np.asarray([np.correlate(x[:-1], x[1:]), np.correlate(x[:-2], x[2:])]) / N
    w_opt = (np.linalg.inv(covariance_matrix_x) @ cross_correlation_vector_dx).squeeze()
    
    p = cross_correlation_vector_dx
    R = covariance_matrix_x
    
    # running iterative optimization
    for index in range(num_steps):
        w_ = np.asarray(w[-1])
        J += [-w_@ p - p.T @ w_.T + w_ @ R @ w_.T]
        w += [list((w[-1] + step_size*(cross_correlation_vector_dx.T - covariance_matrix_x @ w[-1]))[0])]
        
        # convergence condition
        if np.sqrt(np.sum((w[-1]-w_opt)**2)) < epsilon:
            return w_opt, np.asarray(w), index, np.asarray(J)
        
    return w_opt, np.asarray(w), num_steps, np.asarray(J)
```

## d/e) Plot $w(\kappa)$, $v(\kappa)$ and $J(\kappa)$

```python
@interact(step_size=FloatSlider(min=0.1, max=1., step = 0.1, value = 0.3))
def plot_descent(step_size):
    all_a = ((-0.195, 0.95), (-0.9750, 0.95), (-1.5955, 0.95), (-1.9114, 0.95))
    ax = []

    for i, a in enumerate(all_a):
        sigma_u = fun_sigma_u(1, a)
        x = source(a, sigma_u)

        covariance_matrix_x = np.cov((x[::2], x[1::2]))
        eigen_values, _ = np.linalg.eig(covariance_matrix_x)
        print('\nEigenvalue spread is: {}'.format(max(eigen_values)/min(eigen_values)))

        opt_step_size = 2/(max(eigen_values)+min(eigen_values))
        print('Optimal step size: {}'.format(opt_step_size))

        w_opt, w, num_steps, J = grad_descent(x, a, step_size=step_size)
        print('Optimal w: {}\nFound w: {} in {} steps'.format(w_opt, w[-1], num_steps))

        plt.figure(figsize=(10,5))
        
        plt.subplot(121)
        plt.plot(w[:, 0], color='b', label='w_0')
        plt.plot(w[:, 1], color='r', label='w_1')
        plt.legend()
        
        plt.subplot(122)
        plt.plot(J, label='Kostenfunktion J(k)')
        plt.legend()
```
